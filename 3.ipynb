```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("FEV-data-Excel.xlsx")

# Filter out rows with missing values in 'Battery capacity [kWh]' or 'Range (WLTP) [km]'
df = df.dropna(subset=['Battery capacity [kWh]', 'Range (WLTP) [km]'])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Battery capacity [kWh]'], df['Range (WLTP) [km]'], color='blue', alpha=0.7)
plt.title("Relationship Between Battery Capacity and Range", fontsize=14)
plt.xlabel("Battery Capacity [kWh]", fontsize=12)
plt.ylabel("Range (WLTP) [km]", fontsize=12)
plt.grid(True)

# Add a trendline
z = np.polyfit(df['Battery capacity [kWh]'], df['Range (WLTP) [km]'], 1)
p = np.poly1d(z)
plt.plot(df['Battery capacity [kWh]'], p(df['Battery capacity [kWh]']), color='red', linestyle='--', label="Trendline")

# Show correlation coefficient
correlation = df['Battery capacity [kWh]'].corr(df['Range (WLTP) [km]'])
plt.text(20, 600, f"Correlation Coefficient: {correlation:.2f}", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.legend()
plt.show()
```

```python
# The scatter plot shows a clear positive correlation between battery capacity and range.
# As battery capacity increases, the range tends to increase as well.This means that increasing the battery size generally leads to a longer driving range.
# Customers prioritizing long-range EVs should consider models with larger batteries.
# Manufacturers aiming to improve range should focus on increasing battery capacity
```
