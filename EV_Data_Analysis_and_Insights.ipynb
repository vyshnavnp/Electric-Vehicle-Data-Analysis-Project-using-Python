# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("FEV-data-Excel.xlsx")
df.head()
```

```python
df.info()
```

```python
#columns with null values
df.columns[df.isnull().sum() > 0].tolist()
```

```python
df.describe()
```

```python
#A customer has a budget of 350,000 PLN and wants an EV with a minimum range of 400 km

budget=350000
min_range=400

filtered_evs = df[(df['Minimal price (gross) [PLN]'] <= budget) & 
                   (df['Range (WLTP) [km]'] >= min_range)]

print("\nFiltered EVs that meet the criteria:")
filtered_evs = filtered_evs[['Car full name', 'Make', 'Minimal price (gross) [PLN]', 'Range (WLTP) [km]', 'Battery capacity [kWh]']]
filtered_evs
```

#### The filter criteria of a budget of 350,000 PLN and a minimum range of 400 km helped identify a subset of EVs that meet the customer's financial and performance requirements.
#### These results provide a curated list, simplifying the customer's decision-making process for long-range EVs within their budget.
#### There are 12 possible options under this category


```python
#grouping by make
grouped_df = filtered_evs.groupby('Make', group_keys=True).apply(lambda x: x,include_groups=False)
grouped_df
```

#### Grouping by manufacturer revealed which brands offer the most options under the customer's budget and range criteria.
#### This step can highlight dominant players in the EV market segment and help the customer choose based on brand preference.

```python
#Analysing average battery capacity for each manufacturer

average_battery_capacity = (
    filtered_evs.groupby('Make')['Battery capacity [kWh]']
    .mean().round(2)
    .reset_index(name='Average Battery Capacity [kWh]')
)
print("\nAverage Battery Capacity for each Manufacturer:")
average_battery_capacity
```

#### Calculating the average battery capacity for each manufacturer provided insights into which brands invest in larger battery capacities.
#### This is crucial since a larger battery often correlates with a longer driving range and better vehicle performance.
#### We found out that the maker Audi provides the best battery life.

## A closer look at Energy Consumption

```python
#checking for null values
df['mean - Energy consumption [kWh/100 km]'].isnull().sum()
```

```python
#removing rows with null values
df_energy_cleaned = df.dropna(subset=['mean - Energy consumption [kWh/100 km]'])
new_df = df_energy_cleaned[['Make', 'Model', 'mean - Energy consumption [kWh/100 km]']]
new_df.head()
```

```python
#mean energy consumption 
mean_energy = round(new_df['mean - Energy consumption [kWh/100 km]'].mean(),2)
std_energy = round(new_df['mean - Energy consumption [kWh/100 km]'].std(),2)
print(f"mean_energy: {mean_energy}\nstd_energy: {std_energy}")
```

```python

sns.boxplot(
    x = new_df['mean - Energy consumption [kWh/100 km]']
)
plt.xlabel('Energy consumption [kWh/100 km]')
plt.title('Distribution of Mean Energy Consumption')
plt.show()
```

```python
Q1 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.25)
Q3 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df['mean - Energy consumption [kWh/100 km]'] < lower_bound) |
    (df['mean - Energy consumption [kWh/100 km]'] > upper_bound)
]

print(outliers[['Car full name', 'Make', 'mean - Energy consumption [kWh/100 km]']])
```

#### Based on both statistical methods, it can be concluded that the dataset does not contain any outliers in the mean - Energy consumption [kWh/100 km] column. 
#### This suggests that the energy consumption values are relatively consistent across the different electric vehicle models included in the dataset

### Analysing relationship between battery capacity and range

```python
plt.scatter(df['Battery capacity [kWh]'],df['Range (WLTP) [km]'])
plt.title("Relationship Between Battery Capacity and Range")
plt.xlabel("Battery Capacity [kWh]", fontsize=12)
plt.ylabel("Range (WLTP) [km]", fontsize=12)
plt.grid(True)
```

```python
corr_matrix = df[['Battery capacity [kWh]','Range (WLTP) [km]']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Numeric Features", fontsize=14)
plt.show()
```

#### The scatter plot and correlation matrix shows a clear positive correlation between battery capacity and range.
#### As battery capacity increases, the range tends to increase as well.This means that increasing the battery size generally leads to a longer driving range.
#### Customers prioritizing long-range EVs should consider models with larger batteries.
#### Manufacturers aiming to improve range should focus on increasing battery capacity
